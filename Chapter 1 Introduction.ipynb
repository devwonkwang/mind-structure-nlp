# 1장 서론: 마음의 구조와 자연어 처리

## 1. 자연어 처리와 마음의 구조

안녕하세요 조원광입니다. 저를 잘 모르시는 분이 대부분일 거라 짐작합니다. 강의하는 입장으로 여러분을 만났으니 공부한 것 위주로 간략히 설명하자면, 저는 사회학으로 학부/석사/박사를 마쳤습니다. 연구하다가 필요성을 느껴 컴퓨터 과학 석사 학위로 추가로 취득하였습니다. 컴퓨터 과학의 여러 아이디어를 활용하여 여러 영역에서 마음의 구조를 연구하고 그런 방법을 공유하는 것을 업이자 즐거움으로 삼고 있습니다.

이 강의는 자연어 처리를 활용한 사회 탐구, 좀 더 정확히 말해서 자연어 처리를 활용한 마음의 구조 탐구 방법에 대한 것입니다. 저는 자연어 처리가 (Natural language processing) 사회 구조 중 하나인 ‘마음의 구조’를 연구하고 탐구하는데 매우 유용하다고 생각합니다. 기본적으로 자연어 처리는 ChatGPT 등에서 볼 수 있듯 인간의 언어 활동을 모방하는 서비스를 만드는데 활용되고 있고 큰 성과를 거두고 있습니다. 저는 이 기술들이 연구용으로도 충분히 활용될 수 있지만, 그 잠재성이 아직은 충분히 부각되지 않았다고 생각합니다. 이 강의는 이 잠재성을 나름대로 조명하고 소개하기 위함입니다. 

우선 ‘마음의 구조’가 무엇인지 설명해야 겠네요. 제가 말한 **마음의 구조는 우리 인간의 사고와 감정을 특정한 방향과 강도로 형성하는 무의식의 조건과 힘의 장입니다.** 다소 추상적으로 들릴 수 있는데, 여러 학자들이 제시한 사례와 이론을 참조하면 조금 더 느낌이 오실 겁니다. 

부르디외라는 사회학자가 있습니다. 그는 계급에 따라 취향과 미학적 감각이 다르다는 주장으로 유명하지요. 상류층은 클래식 음악 같은 고급 예술을 선호하고 그것을 아름답다 느끼는 반면 노동 계급은 댄스 음악을 즐긴다는 식입니다. 그는 이런 집단적 취향 차이가 특정 계급이 다른 계급과 자신을 구분하는 과정이자 구분하는 힘의 한 측면이라고 지적합니다. 미적 판단이라는 정신 작동이 온전히 개인의 선택이 아니라 집단과 집단간 역학에 의해 형성되고 유도되는 면이 있음을 지적한 것이지요. 저는 이런 것이 일종의 마음의 구조라고 생각합니다. 즉 외부의 여러 힘에 의해 만들어진, 인간이 생각하고 판단하고 느끼는 무의식적 틀이 마음의 구조라는 말입니다. 

이런 마음의 구조는 고정된 것이 아니라 변화합니다. 시간에 따른 변화가 이를 잘 보여줍니다. 푸코라는 사상가이자 철학자가 있습니다. 그는 사람들이 '형벌'을 떠올리고 실행할 때 그것의 전제와 목표 그리고 기술이 시대마다 극명히 달라짐을 지적한 바 있습니다. 중세에는 '화려한 신체형', 즉 공개적이며 잔인한 처벌이 형벌의 모습이었습니다. 군주의 권위에 도전한 '적'에게 '복수'하는 것이 형벌의 주된 골자였습니다. 반면 근대에는 '감옥에서의 훈련과 교화', 즉 감옥이라는 상대적으로 잘 보이지 않는 곳에서 다시 노동하는 시민이 되도록 반복 훈련하는 것이 형벌의 주된 모습이 됩니다. 이것은 우리가 당연하게 여기는 사회의 모습과 국가의 역할에 대한 무의식적 판단이 시대마다 다름을 보여줍니다. 

시간 뿐만 아니라 공간과 문화에 따라 마음의 구조와 정신 작용이 체계적으로 다르기도 합니다. 여러 사회 심리학자들은 문화권에 따라 귀인 패턴이 (attribution pattern) 다름을 지적합니다. 예를 들어 리처드 니스벳은 어떤 사건이 발생했을 때 그 사건의 원인을 개인과 구조에 모두 돌릴 수 있는 상황에서, 동양에서는 사회나 맥락에 귀인을 하고, 서양에서는 개인에게 귀인을 하는 경향이 있음을 지적합니다. 우리가 어떤 일의 원인을 떠올릴 때 무의식적 경향이 문화권에 따라 달라진다는 점을 지적한 것이지요. 

요컨대 많은 이들이 우리의 마음에서 일어나는 여러 일, 예를 들어 판단, 느낌, 감정 등은 오롯이 개인적인 일이 아님을 지적했습니다. 그것은 시대와 공간에 영향을 받고, 내가 소속된 집단과 계급/계층에 영향을 받습니다. 외부의 여러 요소들은 우리 마음에 길을 내고 도랑을 팝니다. 그리고 우리의 마음은 그 도랑을 따라 그 방향으로 흐르면서 커지기도 하고 작아지기도 하죠. 

이런 현상이 있다면, 마음의 구조는 중요한 연구 대상이자 개입 대상입니다. 그것이 개인의 행동은 물론 집단의 갈등이나 협력 같은 인간 상호 작용 전반에 영향을 미치고 우리의 정신적 안녕과도 긴밀히 관련되어 있을 것이기 때문입니다. 개인과 집단들이 보유한 마음의 구조의 실태와 형태를 파악하는 것, 그것에 영향을 미치는 요인들을 분석하는 것, 이것을 변형하거나 바꾸기 위한 방법을 탐색하는 것, 모두모두 중요한 일이 됩니다. 실제로 많은 이들이 이런 마음의 구조를 탐구해왔습니다. 물론 지금 제가 사용하는 것과 동일한 단어나 개념으로 표현하지는 않았지만 말이지요. 예를 들어 '담배를 피게 되는 이유는 무엇일까?' '왜 특정 집단에서 흡연율이 더 높을까?' '왜 담배를 피면서 그 위험에 대해 크게 고려하지 않을까?' 이런 질문은 보건의료계에서 흔하게 볼 수 있는 편인데, 건강 행동 이면의 마음의 구조에 대해 던지는 질문인 셈입니다. 

이제껏 마음의 구조를 연구하는 다양한 방법이 있고 제안되어 왔습니다. 가장 대표적인 것인 설문조사입니다. 직접 당신의 마음이 어떠한지, 즉 어떤 것에 대한 의견이나 감정이 어떠한지 물어보는 것입니다. 조사 대상이 되는 표본을 잘 만들면, 작은 숫자의 표본으로도 큰 집단을 요약하고 대표하는 결과를 얻어낼 수 있다는 점에서, 대체 불가능한 장점을 가진 방법입니다. 잘 훈련된 학자가 여러 문헌을 보고 마음의 구조를 추론해 내는 방법도 있습니다. 앞서 언급한 푸코가 사용한 방법이기도 하죠. 아니면 마음의 구조를 볼 수 있는 행동이나 표현을 실험으로 연구할 수도 있을 것입니다. 

저는 **자연어 처리가 마음의 구조를 탐구한다는 목표에서 기존의 방법들을 보완할 수 있는 면을 가진다고 생각합니다.** 자연어 처리는 자연어, 즉 인간들이 쓰는 자생적으로 형성된 언어를 컴퓨터가 접근하여 처리할 수 있도록 하는 기술 일반을 의미합니다. 이는 다양한 목적으로 쓰일 수 있는데, 자연어를 컴퓨터로 처리하여 거기에 있는 여러 정보를 추출하는 것도 중요한 목적 중 하나입니다. 그리고 자연어 처리 기술과 적합한 인간 언어 데이터가 결합하면 마음의 구조를 조사하고 탐구하는 좋은 방법이 됩니다. 왜냐하면 **앞서 말씀드린 마음의 구조가 정말 있다면, 그것은 인간의 언어 활동에 흔적을 남기기 때문입니다.** 마음의 구조는 여러 가지로 표현될 테지만, 어떤 사람의 생각과 감정을 표현한 언어에도 분명 특이한 흔적을 남길 것입니다. 예를 들어 앞서 말한 귀인 패턴이 집단과 사회를 향해 있는 사람의 경우, 어떤 사건에 대해 서술할 때 자신이 소속된 집단과 사회를 많이 언급하고 자신의 행동과 선택이 어쩔 수 없었음을 강조할 것입니다. 그렇기에 인간의 언어 활동에 있는 특징을 세밀하게 포착할 수 있다면, 거꾸로 마음의 구조를 추적할 수 있는 것이지요. 게다가 자연어 처리를 활용한 방법은 기존의 여러 방법이 부딪혔던 여러 문제를 극복할 잠재력도 가집니다. 

어떤 문제를 극복할 잠재력이 있을까요? 먼저 분석과 연구를 통해 발견하는 것들이 연구자가 미리 마련한 틀에 상대적으로 덜 영향을 받는다는 장점이 있습니다. 설문조사나 각종 실험 방법은 조사자가 미리 설정한 틀 안에서 움직입니다. 설문조사는 조사자가 설정한 어떤 측면에 대한 태도나 의견을 묻습니다. 질문 형태에 따라 다르지만 답변의 범위도 조사자가 설정해 둔 경우가 많죠. 실험 역시 조사자가 미리 설정한 측면의 행동이나 생각을 관찰합니다. 이런 접근은 연구의 초점을 분명히 한다는 점에서는 장점이 있지만, 만약 조사자 혹은 연구자가 탐구 대상 중 중요한 측면을 빠트릴 수 있다는 위험이 있습니다. 예를 들어 식당에 대한 여러 면을 조사하면서 맛이나 서비스에 대해서는 질문하고 조사하지만, 그 식당에 접근성은 질문할 생각을 못할 수도 있습니다. 그런데 만약 사람들이 자발적으로 남긴 식당에 대한 리뷰를 자연어 처리를 통해 분석한다면, 사람들이 중요하다고 생각해서 남긴 여러 자발적 메시지에서 정보를 뽑아내는 셈이므로 연구자가 미리 설정한 틀에 한정되지 않는 다양한 정보를 얻을 수 있습니다. 

타인을 크게 고려하지 않은 응답이나 태도를 조사할 수 있다는 것도 중요합니다. 설문조사건 실험이건 정도의 차이는 있지만 조사자나 연구자와의 상호작용이 포함됩니다. 그리고 우리는 혼자 있을 때와 다른 사람과 함께 있을 때 감정과 태도를 드러내는 방식이 다릅니다. 막스 베버 (Max Weber) 같은 학자가 지적하듯, 타인을 의식하고 타인을 향해서 행동하고 말하는 것이 인간의 특징이기 때문입니다. 그런데 때로는 사람들이 타인을 고려하지 않은 표현에서 그가 가진 마음의 구조를 잘 볼 수 있기도 합니다. 이른바 '솔직한' 취향이나 태도가 중요한 정보를 가질 수 있다는 말입니다. 그런데 만약 소셜 미디어의 익명 계정에 올라온 글이나 화장실에 적힌 낙서 등을 모아서 컴퓨터로 처리하여 정보를 추출할 수 있다면, 타인을 (적어도 명시적으로는) 고려하지 않은 표현의 특징을 볼 수 있고, 그것이 마음의 구조를 탐색하는데 중요한 참조가 됩니다. 

대량의 자료를 처리할 수 있다는 것 역시 장점입니다. 사실 연구자의 틀에서 자유로운 솔직한 표현을 조사하는 것은 과거부터 다양하게 시도되어 왔습니다. 사람들이 남긴 자발적인 여러 기록을 연구자가 조사하는 경우가 많았습니다. 보건의료 영역에서는 환자들의 일기나 경험담을 수집하여 연구하는 사례를 들 수 있겠네요. 그런데 이런 경우는 대부분 인간 연구자가 자료를 종합하고 거기서 정보를 추론했기에 분석할 수 있는 자료의 규모에 분명한 한계가 있었습니다. 한 사람 혹은 소수의 사람들이 수천 개 이상의 신문기사를 읽어야 한다고 가정해봅시다. 불가능하지는 않겠지만, 너무 많은 시간이 들고 결과로 나온 내용 또한 정확치 않을 수 있습니다. 반면 자연어 처리는 근본적으로 컴퓨터를 활용한 처리입니다. 그래서 인간이 분석하는 것보다 훨씬 더 많은 자료를 처리할 수 있지요. 컴퓨터가 우리보다 기억과 계산은 확실히 더 낫기 때문입니다. 

## 2. 세 개의 모델 혹은 접근 방법

그래서 이 강의는 자연어 처리라고 불리는 영역의 기술과 모델 중 마음의 구조를 탐색하는데 이용할 수 있는 것들을 소개하고 이를 활용하여 실제 분석을 연습하는 것을 내용으로 합니다. 저는 세 가지 카테고리에 해당하는 방법 혹은 모델들을 설명하고 실습을 제시할 것입니다. 

- **네트워크 과학을 활용한 분석**
- **토픽 모델링을 활용한 분석**
- **신경망과 딥러닝을 활용한 분석**

이 세 가지가 제가 소개할 범주들입니다. 사실 최근 자연어 처리라고 하면 마지막 범주인 신경망과 딥러닝을 활용한 접근을 떠올리는 경우가 많습니다. 그도 그럴 것이 그 방법이 성취한 것이 너무 눈부시거든요. ChatGPT를 위시한 수많은 언어 AI를 보십시오. 하지만 '연구'라는 관점에서, 즉 어떤 자료에서 정보를 추출한다는 목적에 비추어보면 꼭 신경망과 딥러닝이 늘 좋은 방법이지는 않습니다. 연구는 결과의 설명 가능성과 해석 가능성을 목표로 할 경우가 적지 않기 때문입니다. 그러려면 모델에서 우리가 중요하게 생각하는 변수나 요인의 영향을 분리해 낼 수 있어야 하는데, 이건 역설적으로 모델이 상대적으로 단순하고 통제 가능할 때 달성되는 경우가 많습니다. 그리고 이런 관점에서는 네트워크 분석이나 토픽 모델링이 오히려 최신의 딥러닝 모델보다 나은 경우도 있지요. 

보다 근본적으로 말해, 이 세 가지 접근은 각자 장단점이 있습니다. 이 분석 접근들은 기본적으로 각자의 고유한 '모델'을 동반하고, 서로 다른 모델은 같은 자료에서 '강조해서 부각하는 정보'가 다릅니다. 모델은 기본적으로 현실의 단순화하여 잘 알아볼 수 있게 부각하는 것입니다. 그런데 이를 위해서는 뭘 단순화하여 부각할지 선택해야 합니다. 여러 모델이 존재하는 이유가 그것입니다. 그리고 제가 앞에 말한 세 가지 접근은 서로 다른 모델을 동원하고 있고, 그래서 언어 자료에서 단순화하여 부각하는 면이 서로 다릅니다. 당연히, 이 중 뭐가 좋으냐는 상황과 목표에 달린 문제가 됩니다. 내가 언어 자료에서 A라는 정보를 포착하고 싶다면, 거기에 적합한 모델과 방법을 써야합니다. 모든 면에서 우월하거나 만능인 모델이 있는게 아니라요. 

간단하게 이 세 가지 접근이 동원하는 모델의 특징을 살펴봅시다. 우리가 앞으로 각 접근 방법에 대해 길게 살펴볼 것인데, 지금 일종의 예고편 같은 느낌으로 봐주시면 됩니다. 그러니 정확히 이해가 안되어도 걱정하지 않으셔도 됩니다 (아마, 쓰는 저도 처음에 이렇게만 봤으면 잘 이해하지 못했을 듯하네요^^;;). 큰 그림을 조망하고 싶으신 분도 계실 수 있고, 저도 가끔은 개별 기법 설명에서 빠져나와서 이것의 핵심 위치가 어디인지 생각해보는 시점을 드리려고 노력할 것인데, 그를 위해 일종의 단서를 제공하는 과정으로 생각해주십시오. 

- 네트워크 과학을 활용한 언어 분석들: 
    - 개요: 언어를 이루는 구성요소가 단어 혹은 토큰(token)이라고 하겠습니다. 이들 사이의 유의미한 관계가 무엇인지 정의 후, 그것의 구조를 단순화하여 계량하는 방법입니다. 그 구조적 특징을 보는 다양한 지표 보유하고 있습니다. 
    - 장점: 시스템을 이루는 구성요소들의 (이 경우는 단어 혹은 토큰이죠) 속성 보다는 그들 사이의 특정 관계 구조를 부각하는데 적절합니다. 
    - 단점: 당연히 거꾸로 구성 요소 속성이 강한 영향력을 가지는 상황을 설명하기 어렵고, 여러 종류의 관계를 동시에 고려하기가 쉽지 않습니다. 예를 들어 단어들의 동시 출현 관계와 근접 출현 관계를 동시에 고려하는게 쉽지 않죠. 

- LDA 이후 토픽 모델링: 
    - 개요: 가상적인 확률적 생성 과정을 통해 문서 집합이 등장 했다고 보고, 잠재적인 확률 구조를 추론하는 방법입니다. 
    - 장점: 문서의 생성을 관장한다고 여길 수 있는 개념, 예를 들어 주제/담론/문화 등을 연구하는데 적절한 면이 있습니다. 문서 생성의 일면을 근사합니다. 
    - 단점: 단어 및 토큰들의 특정 단위 동시 출현만 유의미한 정보로 인정합니다. 단어 사이의 순서/근접 출현 등의 정보가 없어집니다. 
    
- 신경망 / 딥러닝 활용한 언어 모델들:
    - 개요: 신경망 모델은 기본적으로 재료로 입력되는 특성/피처들이 만들어내는 비선형적인 결과 산출 패턴을 세밀하게 근사합니다. 표현력이 매우 큰 모델입니다. 언어 자료 처리를 위해 신경망을 활용하는 언어 모델들은 단어/토큰 간 순서와 연결을 다채롭게 고려할 수 있도록 설계되어 있습니다. 그래서 특정 과제 (다음 단어 예측, 빈 단어 예측 등)를 수행하는데 도움이 되는 단어/토큰 사이의 연결을 포착하여 강조합니다.  
    - 장점: 여러 수준의 토큰 간 연결을 과제에 기반하여 유연하게 종합적으로 활용합니다. 인간이 자각 없이 하던 여러 언어 활동을 높은 수준으로 근사 재현합니다. 특히 만들어진 모델의 구조를 들여다보는게 아니라, 모델의 출력을 활용할 때 유용합니다. 
    - 단점: 왜 그렇게 모델이 만들어졌는지, 혹은 왜 모델이 그렇게 작동하는지 설명하기가 힘듭니다. 모델의 표현력을 높이려고 보니, 모델의 파라미터가 감당불가능하게 많아지고, 이들이 의미하는 바도 모호해집니다. 모델의 해석 가능성이 상당 부분 사라진 셈입니다. 


딥러닝의 시대에 굳이 과거부터 존재했던 이 세 가지 접근 방법을 설명하는 이유는, 이처럼 각 접근 방법 혹은 모델의 장단점이 존재하고, 어떤 하나가 다른 것보다 우월하다고 보기가 어렵기 때문입니다. 마음의 구조를 분석하고 연구하는 입장에서, 지금 자료에서 읽어내고 싶은 정보가 늘 같지 않습니다. 그렇기에 적합한 접근과 모델도 같을 수가 없지요. 

더불어 이 강의가 '계량적 방법에 대한 과대한 기대'를 재고하는 일에도 기여하기를 바랍니다. 언어 자료를 계량적 모델을 동원하여 분석하면 더 과학적이고, 재현 가능하고, 합리적이라는 이미지가 있는 듯합니다. 그렇게 보기 어려운 면이 꽤 있습니다. 계량적 방법을 동원한 언어 분석이 흥미로워 보이는 것은, '강조점'이 명확하기 때문입니다. 언어 자료에서 중요한 측면을 선택하여 나름의 방식으로 강조합니다. 때로는 멋진 시각화가 이어지기도 하죠. 그런데 이 장점은 항상 '무엇인가는 보여주지 않고 배경화한다'는 필연적 결과를 동반하기 마련입니다. 뭔가를 강조하려면, 뭔가는 배경이 되어야 합니다. 예를 들어 책을 읽다가 중요 부분에 밑줄을 치는 건, 밑줄 치지 않는 부분은 대강 읽게 만드는 효과를 만들죠. 이것 자체가 문제라고 보기는 어렵지만, 우리가 인지해야 하는 원리입니다. 우리의 모델은 뭔가를 보여주는 동시에, 뭔가를 가리는 면도 있습니다. 이를 자각하고 모델을 활용하는 태도가 필요하다고 생각합니다.
