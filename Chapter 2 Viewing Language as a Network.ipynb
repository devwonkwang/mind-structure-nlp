## 1. 언어와 네트워크

첫 번째 파트에서는 네트워크 과학을 언어 자료 분석에 적용하여 마음의 구조를 탐구하는 방법을 배울 것입니다. 그리고 이 장에서는 언어를 네트워크로 바라봤을 때 어떤 장점이 있고, 실제 언어 자료를 어떻게 네트워크로 전환시킬 수 있는지 배워볼 것입니다. 

네트워크 과학은 어떤 시스템을 구성하고 있는 구성 요소들의 '관계 구조'를 계량화하여 탐구하는데 특화된 방법입니다. 여러 복잡한 시스템에서는 구성 요소들의 상호작용을 통해 각 구성요소로 환원되기 힘든 현상이 만들어집니다. 이걸 '창발'(emergence) 이라고 부르기도 하죠. 이 창발적 현상을 이해하는데 구성요소들 사이의 상호 작용/상호 관계를 체계적으로 연구하는 것이 필요했고, 다양한 영역의 지적 전통이 합쳐져서 현대적인 네트워크 과학이 만들어졌습니다. 

네트워크 과학을 적용한 방법을 배워보기 전에, 이런 접근이 필요한 이유와 장점을 알아보기 위해 세 가지 질문에 답해보겠습니다. 

- **왜 언어를 네트워크로 봐야 하는가?**

많은 이들이 지적하듯이, 마음의 구조를 포착하기 위해 언어 자료를 분석할 때 네트워크 과학의 개념과 도구가 유용한 경우가 많습니다. 마음의 구조가 단어의 연결 같은 현상을 통해 드러나는 경우가 많고, 네트워크 과학은 그 연결의 특징을 포착할 수 있게 해주기 때문입니다. 가장 간단히 떠올릴 수 있는 현상은 어떤 사람이 지닌 **개념의 연상과 그것의 언어적 표현**입니다. 어떤 사람이 특정한 마음의 구조를 가지고 있다면, 즉 판단과 감정이 무의식적 틀을 갖고 있다면, 어떤 사안이나 대상에 대해 사고하거나 느낄 때 특징적인 개념과 관념의 연상 구조를 가질 것입니다. 예를 들어 경제적 효율을 중시하는 집단에 속해 그것에 영향받아 마음의 구조를 형성한 사람이라면, 새로운 보건 의료 정책안을 볼 때, '소요 예산', '실질 효과', '재원', '평가' 같은 개념이 떠오를 것입니다. 반면 취약계층 보호를 중시하는 집단에 속하고 상응하는 마음의 구조를 가진 사람이라면, '수혜자', '불평등 완화' 같은 개념을 떠올리겠지요. 그리고 이런 개념의 연결은 자연스럽게 그 사람의 언어 활동에 반영될 것입니다. 만약 네트워크 과학으로 이처럼 언어 활동에 표현된 개념의 연결, 단어의 연결을 식별하고 추출할 수 있다면, 마음의 구조를 추적하는데 도움이 될 것입니다. 

나아가 개념 연상에 한정하지 않아도 네트워크 과학은 언어 자료 분석에 매우 유용합니다. 왜냐하면 언어가 만들어내는 의미나 가치는 언어에 존재하는 작은 구성요소들의 연결을 통해 만들어지기 때문입니다. 예를 들어 같은 단어라도, 그것이 무엇과 연결되느냐에 따라 매우 다른 의미나 뉘앙스를 가집니다. '작고 귀여운 개가 지나간다'와 '저 개만도 못한 인간'이라는 문장은 둘 다 '개'라는 단어를 포함하고 있지만, '개'가 만드는 의미나 뉘앙스는 각 문장에서 매우 다릅니다. 근처에 다른 종류의 단어들 혹은 언어 요소들이 존재하기 때문이지요. 어떤 단어가 가지는 의미가 근처 단어에 의해 결정된다는 명제는 NLP에서 언어의 분포 가설 (distributional hypothesis) 라는 명칭으로 많이 지적됩니다. 분포 가설을 언급하지 않더라도, 언어의 의미가 구성요소들의 연결을 통해 만들어진다는 점은 언어학에서는 상식에 가까운 명제입니다. 거슬러 올라가보면 소쉬르 같은 학자도 '차이의 테제'라는 이름으로 기호는 다른 기호와의 차이를 통해 의미를 만든다고 지적했지요. 즉 언어 구성 요소의 연결을 봐야, 그 언어 자료에 존재하는 의미와 뉘앙스를 알 수 있다는 말입니다. 그리고 앞서 말했듯 네트워크 과학에는 작은 구성요소들의 연결의 구조와 특징을 연구하고 분석하는 노하우가 집적되어 있지요. 그렇기에 언어 자료를 통해 마음의 구조를 탐구하기 위해서 네트워크 과학을 활용하는 것이 자연스럽습니다. 예를 들어 특정 집단이 생성한 언어 자료의 주된 의미와 뉘앙스를 네트워크 과학을 활용하여 분석할 수 있을 것이기 때문입니다. 

- **네트워크 과학을 통한 언어 자료 연구가 가지는 장점은 무엇인가?**

물론 네트워크 과학을 활용하는 분석이 NLP를 활용한 자료 분석에서 가장 중심에 있다고 볼 수는 없습니다. 오히려 소수적인 접근이라고 말하는 것이, 주로 일부 사회과학자들이 이런 접근을 시도하는 경우가 많다고 표현하는 것이 적절할 것입니다. 하지만 앞서 말했듯 네트워크 과학을 활용한 분석은 언어 데이터를 '연구'하려는 목적에서는 현재 주류적인 방법, 예를 들어 딥러닝 모델들보다 더 뛰어난 면이 있습니다. 그 이유를 꼽자면 여러 가지입니다.  

첫 번째, 연구 목적으로 언어 데이터를 분석할 때 우리의 질문은 때로는 매우 직접적인 관계와 연결과 관련이 있습니다. 이때는 복잡한 모델보다, 이 직접적인 관계와 연결을 특정하여 다룰 수 있는 방법이 더 좋습니다. 예를 들어 여러 암(cancer)의 이미지가 각각 어떠한지 분석하고 싶다고 가정해봅시다. (암종별 문화적 상징과 이미지는 환자 경험에 중요한 요소가 됩니다) 이런 분석을 위한 질문은 때로는 단순하고 직접적입니다. 예를 들어 '폐암'이라는 단어와 같은 문장에서 가장 많이 함께 등장하는 형용사는? 혹은 ‘폐암’과 3단계 거리 이내에서 가장 많이 등장하는 형용사는? 이라고 질문을 던질 수 있습니다. 나아가 암에 대한 여러 기사들에서 가장 중심적인 개념이 궁금할 수 있습니다. 그 경우 가장 다양한 단어와 함께 출현한 단어가 무엇인가? 라고 질문을 던질 수 있습니다. 이런 질문들에 대답하기 위해서는 '관계'와 '연결'을 직접적으로 측정하고 계량화하는 방법이 필요합니다. 네트워크 과학은 이를 위한 상대적으로 간단하고 직접적인 방법을 제공합니다.

두 번째, 첫 번째에서 언급한 직접성 때문에 생기는 장점인데, 여러 종류의 관계를 구분하여 측정 혹은 연구할 수 있습니다. 구성 요소 사이의 관계란 여러 가지입니다. 단어로 예를 들자면, 어떤 사람이 '단어들이 서로 연결되어 있다 혹은 관계를 가진다'고 말할 때, 그것은 1) 복수의 단어가 같은 구조적 단위 (e.g., 문장, 단락)에 동시에 출현하는 것을 의미할 수도, 2) 근처에 등장하는 것을 의미할 수도 (e.g., 두 단어가 5단계 이내에 존재), 3) 앞서 두 가지 의미에서 두 단어와 연결된 제 3의 단어들이 서로 비슷한 것을 의미할 수도 (e.g., '대통령 (president)'도 '예산', '정책', '국가'라는 단어와 연결되어 있고, '수상(prime minister)'도 '예산', '정책', '국가'라는 단어와 연결되어 있다. 그래서 대통령과 수상이 관련이 있다) 있습니다. 이 각각은 모두 '연결되었다'라는 같은 말로 묘사되는 경우가 많지만, 사실 다른 종류의 연결을 가리키며, 그 효과 역시 다릅니다. 예를 들어 어떤 욕설과 연결된 단어를 1번이나 2번의 의미에서 찾는다면 그 단어들은 욕설의 대상이 될테지만, 3번의 의미에서 찾는다면 그 욕설과 대체 가능한 유사한 욕설이 될 것입니다. 네트워크 과학은 이런 차이나는 관계의 질(quality)을 서로 구분하여 포착하게 만들어줍니다. 

세 번째, 연결과 관계를 언어 외부로 확장하려 할 때 도움이 될 수 있습니다. 사실 언어의 의미나 효과가 언어 구성 요소들 사이의 연결에만 기대는 것은 아닙니다. 언어의 요소와 언어 외적인 요소의 연결이 효과를 만들어내기도 합니다. 예를 들어 제가 같은 말을 해도, 어떤 표정과 상황에서 그 말을 하는지에 따라 전달하는 의미가 달라집니다. '잘했어'라는 똑같은 말이 상황과 표정에 따라 비꼬는 말이 되기도, 격려하는 말이 되기도 하는 것처럼요. 이런 것까지 다루기 위해서는 명시적으로 유의미한 연결을 고려하고 그것으로 분석을 확장할 수 있어야 합니다. 그러기 위해서는 네트워크 과학의 도구가 유용할 수 있습니다. 

- **네트워크 과학만 연결을 고려하는가?**

오해하지 말아야 할 것은 언어가 가진 관계적 정보를 활용하는 것이 네트워크 과학을 활용한 분석에서만 일어나는 일은 아니라는 점입니다. 예를 들어 최근 언어 AI의 핵심 아키텍쳐인 트랜스포머와 셀프 어텐션 역시 토큰 간 연결을 중요하게 취급합니다. 각 토큰에서 일어나는 연산이 다른 토큰과 연결되고 의존한 상태로 일어나도록 한 것이 기술의 핵심이라 봐도 무방합니다. 그렇기에 언어가 가진 관계적 정보를 활용하여 특정 과제에 대한 성능을 높이고 있다고 설명해도 될 것입니다. 다만, 이때 활용되는 관계적 정보가 매우 다양하여, 각 정보가 가진 효과를 분리해서 보는 것에 이들이 아주 적합하다고 보기는 힘듭니다. 정확히 말해, 불가능하진 않겠지만 애초에 그러려고 만들어진 모델들이 아니기에 꽤나 복잡한 과정을 거쳐야 하지요. 예를 들어 GPT사전 모델에서 토큰 연결 정보를 추출하는 것이 불가능하지는 않겠지만, 꽤나 공을 많이 들여야 할 것입니다. 더불어 애초에 그걸 위한 것도 아니기 때문에 해석 가능성도 제한적이지요. 반면 네트워크 과학은 애초에 관계의 종류를 구분하고, 그것을 측정하고 관찰하려는 목적을 가지기에 이런 일을 좀 더 잘 해낼 수 있을 것입니다. 이건 이 강의의 후반부에 나올 신경망/딥러닝에 대한 부분을 공부하고 나면 더 와닿으실 것입니다. 

## 2. 네트워크 과학 용어 및 형식 간단 소개

### 2.1. 네트워크 과학의 기초 용어와 그것이 가치

본격적으로 언어 자료를 네트워크 과학을 통해 분석하기 전에, 네트워크 과학의 기본적인 용어와 자료 형식 그리고 그것의 강점에 대해 살펴봐야 할 듯합니다. 너무 길지 않게 알아본 후, 언어 자료 분석으로 들어가겠습니다. 우선 네트워크란 무엇인가요? 교과서적인 정의는 매우 추상적입니다. 예를 들어 마크 뉴먼(Mark Newman)은 다음과 같이 네트워크를 정의합니다. 

> A network is, in its simplest form, a collection of points joined together in pairs by lines (_Networks_ by Newman)

점들과 그들을 연결하는 선들의 집합이라는 말인데요. 그래서 네트워크를 말할 때 그를 반영한 다음과 같은 이미지를 주로 떠올립니다. 여기서 점은 노드(node)라고 부르고 선은 링크(link)라고 부르는 경우가 많습니다. 정확히 말해 분야에 따라 점은 vertice, point, 선은 edge, line, tie 등으로 다양하게 불리나, 실질적인 의미는 동일합니다. 아래의 코드를 실행시키면 나오는 그림이 네트워크하면 떠오르는 대표적인 이미지이지요. 아래의 파란 동그라미가 노드, 그 노드들 사이의 선이 링크인 셈입니다. 

```python
import networkx as nx
import matplotlib.pyplot as plt

# 그래프 생성
G = nx.karate_club_graph()

# 레이아웃 고정 (seed 부여)
pos = nx.spring_layout(G, seed=42)

# 시각화
plt.figure(figsize=(6, 4))
nx.draw(
    G,
    pos,
    with_labels=False,
    node_size=100
)
plt.show()
```

상당히 많은 대상 혹은 시스템을 이런 네트워크로 표현할 수 있습니다. 점으로 설정한 것과 선으로 설정할 것만 명확하면 되지요. 예를 들어 같은 수업을 수강하면 연결되었다고 가정하고 한 대학교의 학생들을 네트워크로 표현할 수 있을 것입니다. 시/군 단위 지역을 노드로 보고, 지역 사이에 고속도로가 있으면 링크가 있다고 가정한 후 표현할 수도 있겠지요. 하지만 이렇게 표현된 네트워크가 모두 다 유의미하다는 보장은 없습니다. 그래서 유의미한 네트워크를 포착하는 것도 중요한 절차입니다. 

그러면 언제, 왜 네트워크를 통해 시스템을 표현하게 될까요? 원칙적으로 말하자면, 상호작용하는 요소가 많은 복잡한 시스템을 모델링하고 연구할 때 네트워크가 많이 사용됩니다. 어떤 시스템의 작동과 결과를 이해하려고 할 때, 구성요소들의 상호작용과 그 상호작용의 구조가 중요하고, 그걸 네트워크 과학의 언어로 포착하기가 좋다는 말입니다. 언어가 의미를 만들어내는 것도 세부 구성 요소의 복잡한 상호 작용을 통해서이기 때문에 언어를 네트워크로 전환할 가치가 생기는 것이지요. 

이 설명이 좀 추상적일 수도 있을 듯해서, 저에게 익숙한 소셜 네트워크의 사례를 들어보겠습니다. 개인적으로 저는 소셜 네트워크 이론과 분석을 접하면서 네트워크 과학이라는 분야를 알게 되었습니다. 인간 관계를 네트워크로 표현하고 그 효과를 보려는 시도가 과거부터 있었습니다. 사람들의 상호작용이 집단적, 개인적 효과를 내고 그것을 단순히 개인들의 영향력을 더하는 것으로는 설명할 수 없었기에 당연한 접근이라고도 할 수 있습니다. 

간단히 예를 들어보자면 다음과 같습니다. 다음 그림은 직장 내 관계를 보여주는 네트워크라고 하겠습니다. Situation 1과 2에서 A라는 동일한 사람이 위치해 있습니다. 일종의 평행 세계라고 생각합시다^^ 각 노드는 사람이고, 링크는 조언을 주고 받는 관계라고 하죠. A가 situation 1에 있을 때와 situation 2에 있을 때 둘 중 어느 쪽에서 '아이디어가 좋다'는 평을 들을 가능성이 클까요?

늘 동일하다고 보긴 어렵지만, 보통 situation 1에서의 A가 아이디어가 좋다고 평가받을 가능성이 더 크다고 이야기합니다. 양쪽의 차이의 핵심은 뭘까요? situation 1에서는 A의 이웃들이 서로 연결되어 있지 않습니다. 그리고 각자가 다른 집단에 속해 있지요. 반면 situation 2에서는 A의 이웃들이 서로 다 연결되어 있습니다. A의 친구들이 서로 잘 아는 형태인 거죠. 그리고 내 이웃들이 서로 잘 모를 때, 내가 어떤 자원의 흐름을 조절할 수 있는 일종의 매개자 혹은 중개자 (broker) 위치에 놓일 가능성이 더 커집니다. 예를 들어 situation 1에서의 A는 B 그룹에서 나온 아이디어를 듣고, 그걸 D들이 모인 그룹이나 C들이 모인 그룹에 전달할 능력이 생깁니다. 일종의 구조적 위치에서 나오는 능력이죠. C그룹에서 듣고 D로 옮기는 것도 가능할 것입니다. 엄청 똑똑해 보이겠죠. 그런데 situation 2는 그게 불가능합니다. A가 그걸 하려고 한다고 치죠. B그룹에게 들은 이야기를 D그룹에 속한 내 친구인 D1에게 하려고 한다고 칩시다. 그럼 D1은 이렇게 말할지도 모릅니다. '그거 B1한테 나도 들었는데.' 즉 '전달자', '매개자'로서의 역할을 하기 어렵게 됩니다. 

사실 그렇게 가상적인 예도 아닙니다^^ 제가 좋은 사례인데, 강의하고 있는 저는 사회학과에서 학부부터 박사까지 학위를 했죠. 나중에 추가로 컴퓨터 과학을 공부했고, 보건대학원에서 근무하고 있습니다. 각 커뮤니티에 있는 분들은 저를 알고 교류하지만 서로 자세히 알지는 못하죠. 그래서 보건대학원에서 '사회적 자본 (social capital)'이나 푸코의 '담론 (discourse)' 따위에 대해 이야기하면, 매우 흥미롭게 들으십니다. 거꾸로 사회학을 공부하며 만난 지인들에게 '코티솔'과 같은 생체표지자 (biomarker)를 활용한 연구에 대해 이야기하면 마찬가지의 효과가 나는 경우가 많죠. 사실 양 커뮤니티에서 그런 이야기는 좀 일반적임에도 불구하고 말이죠. 어찌보면, 저의 직업적 생존에도 이런 중개자 (broker) 효과가 꽤 있었을 듯합니다. 

로날드 버트 (Ronald Burt) 라는 사회학자는 이런 현상을 '구조적 공백 (structural hole)'이라는 개념으로 설명합니다. 즉 마땅히 있을 법한 연결이 없는 상태에서 중개자의 이익이 커지는 일이 생긴다는 거죠. 이건 구성 요소들 사이의 '관계'가 구성요소들의 속성이나 특징으로 환원되기 어려운 어떤 종류의 효과를 만들어내는지 보여주는 소셜 네트워크적 예시라고 할 수 있습니다. 저 예에서 A라는 사람의 속성이나 특징은 변화하지 않았는데, 관계 구조가 차이를 만들어낸 셈이니까요. 

그리고 언어를 네트워크로 보는 것은, 언어의 구성요소들의 역할/의미/기능 역시 다른 구성요소들 사이의 관계를 통해 변화하기 때문입니다. 그리고 다른 영역에서 정립된 지표나 구조적 특징을 언어 자료에 써볼법한 경우도 적지 않지요. 요컨대, 네트워크 과학은 구성 요소들의 관계를 추상화하고 계량화합니다. 그것은 주로 구성요소들의 상호작용이 만들어내는 비선형적 효과를 포착하기 위함인 경우가 많았습니다. 그리고 그걸 언어에도 적용해볼만한 것이지요. 우리가 언어 자료에서 관심있는 것, 예를 들어 그 자료에 존재하는 주제/의미/뉘앙스/마음의 구조의 흔적 따위가 언어 구성요소들의 상호작용을 통해 만들어지기 때문입니다. 


### 2.2. 네트워크 정보의 저장과 표현 그리고 네트워크의 기본 유형

그러면 네트워크 과학에서 다루는 이런 구성 요소 '사이'에 존재하는 관계적 정보와 연결 정보는 어떻게 저장하고 표현할까요? 위에서는 그림으로 표현하였습니다. 직관적 매력이 있지요. 그런데 늘 그림으로 그릴 수는 없습니다. 무엇보다 네트워크가 커지면, 그림으로 그리는게 사실상 불가능해집니다. 노드가 200개만 넘어가도 보통 그냥 '엄청 복잡하다'는 인상 밖에 안들거든요. 그래서 이 정보를 안정적으로 저장하고 표현할 수 있는 다른 방법이 필요합니다. 여러 가지 방식이 있습니다만, 가장 일반적이고 우리가 이해하기도 쉬운 방법이 행렬(matrix)로 표현하고 저장하는 것입니다. 예를 들어 아래 그림에서 왼쪽과 같이 시각화할 수 있는 네트워크를 오른쪽 같은 행렬로 저장하는 것이죠. 이것을 인접 행렬(adjacency matrix)이라고 합니다.

행렬에 연결 정보를 저장하는 방식은 간단합니다. 행렬의 i행 j열이 있다고 하면, 그곳을 i 번째 노드와 j번째 노드 사이의 정보를 저장하는 공간으로 삼는 것입니다. 간단히 보자면 위의 사례처럼 연결이 있으면 1, 없으면 0으로 삼는 것입니다. 

네트워크의 종류에 따라 행렬의 형태가 조금씩 달라집니다. 예를 들어 위와 같이 연결의 존재 여부만 보고 강도나 방향을 고려하지 않는 무방향/무가중 네트워크(undirected unweighted network) 라고 할 수 있을 것입니다. 여기서 연결의 강도를 고려하는 것을 추가할 수 있습니다. 그냥 연결이 있다 없다가 아니라 강도를 고려할 수 있다는 말입니다. 예를 들어 같은 친구 사이라도 한달에 한 번 보는 사이와 한달에 다섯 번 보는 사이를 다르게 놓을 수 있지요. 이 경우 행렬 성분 (entry)에 0/1이 아니라 다양한 실수를 넣을 수 있습니다. 이를 가중 네트워크(weighted network)라 합니다. 

다음으로 네트워크를 이루는 연결에 방향 개념을 넣을 수도 있습니다. A와 B가 연결되었다고 보는 것에서 더 나아가, A에서 B를 향해 링크가 뻗어나갔다고 보는 셈이지요. A-B와 A->B 의 차이랄까요. 친구 사이로 예를 들자면 A와 B가 친구라고 말하는 것과 A가 B를 친구로 생각한다는 것은 다르니까요. 행렬에서 이를 나타내는 방법은 i행 j열과 j행 i열을 다르게 취급하는 것입니다. 무방향 네트워크 (undirected network)에서 둘은 동일합니다. 즉 무방향 네트워크는 대칭 행렬로 표현됩니다. 반면 방향이 있는 네트워크 (directed network)에서 둘은 다르게 표현될 수 있습니다. i행 j열은 i로부터 (source) j를 향하는 (target) 링크의 존재 여부를 표현하게 만드는 것이지요. 거꾸로 j행 i열에 i가 source고 j가 target인 링크를 표시하는 경우도 있습니다. 마크 뉴먼에 따르면 이것이 계산상 더 좋다고 하네요. 제 경험상 어느쪽이든 일관되게 유지하고 설명하기만 하면 큰 문제 없는 듯합니다.

이중 모드 네트워크(Bipartite network)도 네트워크의 중요한 유형 중 하나입니다. 이중 모드 네트워크는 노드들을 두 집단으로 나눌 수 있고, 각 집단에 속한 노드들이 다른 집단에 속한 노드들과만 연결을 형성하는 네트워크를 뜻합니다. 예를 들어 학생들의 동아리 소속을 네트워크로 표현한다고 해보죠. 그 경우 학생과 동아리 사이에만 연결이 존재할 수 있습니다. 학생이 다른 학생에게 소속될 수도, 동아리가 다른 동아리에 소속될 수도 없으니까요. 소셜 네트워크 분석에서는 이를 2-모드 네트워크(2-mode network)라 부르기도 합니다. 이런 이중 모드 네트워크의 경우 행과 열이 서로 다른 종류의 노드의 연결 정보를 가리키기 마련입니다. 그래서 행과 열의 길이가 같은 정방 행렬(square matrix)로는 표현하기 어렵습니다. 대신 행과 열의 길이다 다른 직사각 행렬(rectangular matrix)로 표현합니다. 그리고 이를 발생 행렬(incidence matrix)이라 부릅니다. 아래와 같은 형태라고 할 수 있지요.

이렇게 계산을 하면, 마지막에 나오는 행렬은 사람들이 동아리에 몇 번이나 동시 소속되었는지 알리는 정보를 표현합니다. 예를 들어 마지막 행렬의 1행 2열의 성분은 1번 사람과 2번 사람이 몇 개의 동아리에 '동시 소속' 되었는지 알려주는 정보가 됩니다. 왜 이게 되는 걸까요? 간단합니다. 행렬의 곱이 어떻게 계산되는지만 생각해보면 됩니다. 마지막 행렬의 1행 2열에 위치한 성분은 어떻게 계산될까요? 첫번째 행렬의 1행과 두번째 행렬의 2열의 정보를 dot product, 즉 '순서에 따라 짝을 맞춰서 곱한 값'이 됩니다. 그런데 첫번째 행렬의 1행은 어떤 정보인가요? 그건 1번 사람의 동아리 소속 정보입니다. 두번째 행렬의 2열은요? 그건 2번 사람의 동아리 소속 정보입니다. 이걸 짝 맞춰서 곱하면, 동시에 1인 칸만 1이 되어 살아남어서 최종 결과에 더해집니다. 실제 계산을 보면 첫번째 행렬의 1행은 (1,0,1,0) 이고 두 번째 행렬의 2열은 (1,0,0,0)인데, 이걸 짝맞춰 곱해서 더하면, $1\times1+0\times0+1\times0+0\times0$ 이 되죠. 보다시피, 둘 다 첫번째에서 1이라서 그 값만 0이 아닌 값을 가지게 되고 그래서 결과에 더해집니다. 그런데 첫번째 값이 둘 다 1이라는 말은 무슨 말일까요? 그건 1번 사람도, 2번 사람도 첫번째 동아리인 football에 함께 소속되었다는 말이죠. 그러니 이렇게 계산하면 '동시 소속'을 말해주는 행렬이 나오는 것입니다. 

사실 이건 언어 네트워크 분석에서 매우 중요한 테크닉입니다. 이게 주로 많이 활용되는 '토큰 간 공동 출현 네트워크'를 만드는 기본적인 테크닉이거든요. 사실 우리가 문서 집합 같은 언어 자료를 얻으면, 가장 쉽게 파악할 수 있는 정보가 어떤 단어 혹은 토큰이 어디에 나왔는지 입니다. 그걸 소속을 의미하는 이중 모드 네트워크로 표현하는건 아주 쉽죠. 우리가 위에서 클럽으로 표현한걸 문서 번호나 단락 번호로 바꾸고, 사람을 단어 혹은 토큰으로 여기면 그만입니다. 그런데 우리가 흔히 활용하는 연결 가정이 '같은 문서 혹은 단락에 출현하면 연결되었다'는 것인데, 이건 정확히 위의 테크닉으로 접근할 수 있는 연결 정의입니다. 

이것 말고도 네트워크 과학에 중요한 내용이 많습니다. 네트워크의 종류나, 정보를 저장하는 형식에 대해 설명하자면 끝이 없죠. 하지만 제 생각에 이 정도만 알아도 언어 분석에 적용하기는 충분한 듯합니다. 나머지는 그때그때 필요하면 설명 드리겠습니다. 아래의 핵심만 기억합시다. 

1. 네트워크 정보는 행렬로 저장할 수 있다. 
2. 네트워크 유형에 따라 행렬의 형태가 달라질 수 있다.

### 3.2. 동시 출현 네트워크 만들기 (co-appearance network)

네트워크를 만들려면 일단 연결의 기준이 있어야 합니다. 보통 많이 선택되는 것이 특정한 단위에서 동시에 출현했느냐 여부입니다. 예를 들어 문장에 함께 출현하면 두 토큰이 연결되어 있다고 가정하는 것이죠. 한 문장에 쓰였다면 서로 연결되어 어떤 의미를 만들어낼거라고 짐작할 수 있으니 꽤 타당한 가정이라고 생각합니다. 2장에서는 이것만 실습해보겠습니다. 

위와 같은 자료가 주어졌을 때 어떻게 동시 출현 네트워크(co-appearance network)를 만들 수 있을까요? 코드를 실행하기 전에 대략적인 순서를 말로 설명하면 다음과 같습니다. 

1. 토큰 사전을 만듭니다. 여기서 토큰 사전이란 토큰에 인덱스를 부여하여 서로 왔다갔다 할 수 있도록 하는 자료를 의미합니다. 더불어 토큰 빈도도 계산해 둘 것입니다. 빈도 기반으로 필터링을 하는 경우가 있거든요. 
2. 문장에도 인덱스를 부여합니다. 
3. 토큰 인덱스와 문장 인덱스에 근거하여 자료를 행렬로 전환합니다. 희소 행렬(sparsematrix) 같은 자료 구조를 이용합니다. 이건 대용량 데이터에서도 작업을 원활하게 하기 위함입니다.
4. 해당 행렬 사이의 곱을 이용해서 토큰 간 공출현 정보를 만듭니다. 앞서 언급한 이중 모드 네트워크의 변환을 활용하는 것입니다. 
5. 만들어진 결과를 다듬어서 네트워크 패키지에 넣습니다. 이 강의에서는 networkx를 사용할 것입니다.

한 번 하나하나 따라가 봅시다. 

#### 3.2.1. 토큰 사전 만들기

```python
## importing library

from collections import Counter, defaultdict
from itertools import combinations
import numpy as np
import scipy.sparse as sp
import networkx as nx
```

```python
## 사전 만드는 함수 설정

def build_vocab(corpus, min_df=1, max_df=1.0, top_k=None, keep_order=False):
    """
    corpus: List[List[str]] / 우리 코퍼스 형태
    min_df: 최소 문장 빈도(문장 수 기준)
    max_df: 최대 문장 비율 (0~1). 1.0이면 제한 없음. / 모든 문서 혹은 문장에 등장하는 경우를 제어하기 위한 옵션
    top_k: 상위 빈도 토큰 수 제한(문장 빈도 기준). None이면 제한 없음. 
    keep_order: True면 등장 순서 유지(동점 시) 위해 안정 정렬
    
    return: token2id(dict), id2token(list), doc_freq(Counter)
    """
    S = len(corpus)
    doc_freq = Counter()
    for sent in corpus:
        doc_freq.update(set(sent))  
        # 한 문장에서는 여러 번 등장해도 한 번만 등장하는 것으로 처리하고 있습니다 / 기본적으로 이 부분을 통해 단어별 빈도가 계산됩니다

    # max_df 정의
    max_df_count = int(np.ceil(max_df * S))

    # 필터링
    items = [(tok, df) for tok, df in doc_freq.items()
             if df >= min_df and df <= max_df_count]

    
    # 상위 top_k
    # 기본 정렬 기준: df 내림차순 (즉 많은 문서에 출현한 토큰이 앞쪽 인덱스를 받음), 
    # 동률 시 토큰명
    # keep_order면 코퍼스에서 먼저 등장한 token이 앞쪽으로
    
    if keep_order:
        first_pos = {}
        for i, sent in enumerate(corpus):
            for t in sent:
                if t not in first_pos:
                    first_pos[t] = i
        items.sort(key=lambda x: (-x[1], first_pos.get(x[0], 10**9)))
    else:
        items.sort(key=lambda x: (-x[1], x[0]))

    if top_k is not None:
        items = items[:top_k]

    id2token = [tok for tok, _ in items]
    token2id = {tok: i for i, tok in enumerate(id2token)}
    
    # doc_freq는 선택된 단어만 남겨 반환합니다. 
    doc_freq = Counter({tok: df for tok, df in items})
    return token2id, id2token, doc_freq
```

```python
token2id, id2token, df = build_vocab(
    corpus_tokens, 
    min_df=1,   # 너무 희귀한 토큰 제거 기준 / 지금처럼 놔두면 1번만 등장한 것도 포함하겠다는 것
    max_df=1, # 너무 흔한 토큰 제거 비율 / 지금처럼 1로 놔두면 모든 문서에 등장해도 포함하겠다는 것
    top_k=None, # 상위 N개 제한 없으면 None
    keep_order=True
)
```

#### 3.2.2. 문장과 토큰 모두 인덱스로 전환하기

```python
def sentences_to_indices(corpus, token2id, binary=True):
    """
    binary=True면 문장 내 중복 제거(set) 한 문장에서 1회 등장으로 처리
    """
    idx_sents = []
    for sent in corpus:
        indices = [token2id[t] for t in sent if t in token2id]
        if binary:
            indices = sorted(set(indices))
        idx_sents.append(indices)
    return idx_sents
```

```python
## 우리에게 주어진 코퍼스를 index 리스트로 변환하기

idx_sents = sentences_to_indices(corpus_tokens, token2id, binary=True)
```

#### 3.2.3. 행렬로 표현하기

```python
def build_incidence_matrix(idx_sents, V, binary=True):
    """
    형태: (S x V)  문장×어휘
    binary=True면 출현 어휘마다 1을 입력 , False면 문장 내 출현 횟수 count해서 입력
    """
    rows, cols, data = [], [], []
    for s, idxs in enumerate(idx_sents):
        if binary:
            for t in idxs:
                rows.append(s); cols.append(t); data.append(1)
        else:
            # 문장 내 빈도로 기록
            c = Counter(idxs)
            for t, cnt in c.items():
                rows.append(s); cols.append(t); data.append(cnt)
    S = len(idx_sents)
    A = sp.csr_matrix((data, (rows, cols)), shape=(S, V), dtype=np.float64)
    
    ## CSR_matrix는 scipy의 sparsematrix 자료 구조 중 하나입니다.
    ## 행 단위로 데이터와 인덱스를 압축 저장하여 행 인덱싱이 빠르고 행렬곱에 최적화되었다고 알려졌습니다. 그래서 이걸 씁니다. 
    
    return A
```

```python
## 문장 혹은 문서를 행으로, 토큰을 (정확히 말해 토큰 인덱스를) 열로 놓은 행렬을 구현합니다. 

V = len(id2token)
A = build_incidence_matrix(idx_sents, V, binary=True)

print(A)
```

#### 3.2.4. 동시 출현 정보를 담은 행렬로 전환

```python
def cooccurrence_from_A(A, zero_diagonal=True):
    """
    A: (S x V) CSR
    return: W (V x V) CSR
    """
    W = A.T @ A
    if zero_diagonal: # 기본적으로 대각 성분은 해당 token의 출현 횟수가 됩니다. 근데 이건 필요가 없죠. 그래서 주로 지웁니다
        W = W.tolil()
        W.setdiag(0)
        W = W.tocsr()
        W.eliminate_zeros()
    return W
```

```python
W = cooccurrence_from_A(A, zero_diagonal=True) 
print(W)
```

#### 3.2.5. networkx에 넣기

```python
def to_networkx(W, id2token, weight_threshold=0.0):
    """
    W: (V x V) -> 차례대로 형성되어 왔다면, 원리 상 대칭 행렬이 되어 있습니다
    weight_threshold: 링크 가중치 하한 (이상만 보존)
    이런 방식 말고, networkx의 from_scipy_sparse_array 등의 함수를 쓰는 방법도 존재할 수 있습니다.
    """
    W = sp.triu(W, k=1).tocoo()  # 중복 방지(상삼각 행렬 성분남 남기기) 양쪽에 정보가 있기 때문에 한 쪽은 지웁니다
    G = nx.Graph()
    for i, tok in enumerate(id2token):
        G.add_node(tok)
    for i, j, w in zip(W.row, W.col, W.data):
        if w >= weight_threshold:
            G.add_edge(id2token[i], id2token[j], weight=float(w))
    return G
```

```python
G_count = to_networkx(W, id2token, weight_threshold=1)       # raw count 기준

print(G_count.nodes(), "\n")
print(G_count.edges(data = True), "\n")
```

#### 3.2.6. 간단한 시각화

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 8))

# spring layout (force-directed)
pos = nx.spring_layout(
    G_count,
    seed=42,
    k=1, iterations = 200)      # 기본 거리 키우기 (노드 수 많을수록 크게 잡아야 함)

# 네트워크 시각화는 반드시 노드 배치의 논리를 결정하는 레이아웃이 필요합니다

# 기본 그리기 (노드 + 엣지 + 레이블)
nx.draw(G_count, pos,
        with_labels=True,
        font_family="Malgun Gothic",  # 한글 노드 깨짐 방지
        node_color="lightblue",
        node_size=100,
        font_size=8,
        edge_color="gray",
        alpha=0.8)

plt.title("Co-occurrence Network (spring layout)")
plt.show()
```

### 3.3. 윈도우 기반 근접 출현 네트워크 (window based co-occurrence network) 만들기

위의 실습은 재미있으셨나요? 사실 언어 자료를 네트워크로 전환시키는 방식은 다양합니다. 연결에 대한 정의가 다양할 수 있기 때문입니다. 제 개인적 감으로는 위와 같이 특정 단위를 기준으로 한 토큰 동시 출현을 연결로 가정하는 경우가 가장 흔합니다. 우리가 방금 한 것처럼 '문장'에 동시에 출현하면 연결되었다고 보는거죠. 

그것 만큼이나 자주 쓰이는 연결 가정이 특정 거리 이내에서 함께 출현하면 연결되어 있다고 보는 가정입니다. 예를 들어 '마음의 구조와 자연어 처리를 쓴 사람은 **정말** 착하고 잘생긴 사람이라고 들었다.'라는 문장이 있다고 하죠. 그 중에 제가 굵게 표시해 둔 '정말'이라는 단어와 2단계 이내에 출현하면 그 단어와 연결되었다고 가정해봅시다. 그러면 '정말'은 '쓴', '사람', '착하고', '잘생긴'과는 연결되어 있지만, '마음의', '구조와', '자연어' 등과는 연결되어 있지 않습니다. 그들과는 같은 문장이기는 하지만 거리가 2를 넘어가거든요. 이 경우 우리는 window 2를 설정하여 '정말'과 근접 출현 네트워크를 형성하는 토큰을 찾은 셈입니다. 이걸 모든 토큰에 대해 반복하면, 그게 바로 윈도우 기반 근접 출현 네트워크인 것이지요. 

그렇기에 동일한 자료에서도 다른 종류의 연결 정의를 동원하면 다른 네트워크를 만들 수 있습니다. 그 중 뭐가 타당하냐, 혹은 뭐가 언어 작동에서 유의미한 정보를 주느냐는 상황에 따라 다릅니다. 앞서 말했듯 네트워크라고 무조건 정보를 주는게 아니라, '유의미한 네트워크'가 있기 마련이거든요. 여기서 분석자의 이론적 관점과 시각이 필요한 것이지요. 다만 여기서 이에 대해 길게 이야기하기보다는, 일단 윈도우 기반 근접 출현 네트워크를 만들어봅시다. 이런 논의를 할 기회가 또 있을 것입니다. 

자 그러면 어떻게 이런 방식의 연결을 네트워크로 표현할 수 있을까요? 앞서 했던 작업과 크게 다르지 않습니다. 대략 진행 순서는 다음과 같습니다. 

1. 사전을 만든다 (토큰 <-> 인덱스 체계를 만들고, 토큰 빈도도 필요하면 계산)
2. 코퍼스를 살펴보면서 윈도우 한계 내에서 출현한 토큰간 짝 (pair) 데이터를 만든다 (주로 for 구문 안에서 인덱스 한계를 조정해가면서 진행합니다)
3. 토큰 짝 정보를 행렬로 전환한다
4. networkx등의 네트워크 라이브러리로 전환한다


사실 여기서 굳이 행렬로 전환하지 않고, edge list 혹은 link list라는 형태의 정보를 산출하고 그걸 networkx에 넣는 방법이 더 간단할 공산이 큽니다. 그런데 우리가 앞서 이 개념들을 배우지 않았으니, 일단 행렬로 전환한 후 넣는 방식을 취해보겠습니다. 

#### 3.3.1. 각종 헬퍼 함수 설정

앞에서 순서대로 한 번 해봤으니, 이번에는 위의 절차를 따르기 위한 함수들을 미리 설정하고 진행하겠습니다. 아래와 같습니다. 

```python
from collections import Counter
import numpy as np
import scipy.sparse as sp
import networkx as nx


## 1. 사전 만드는 함수 설정 (앞서 사용 했던거 약간 간단히)

def build_vocab(corpus, min_tf=1):
    tf = Counter()
    for sent in corpus:
        tf.update(sent) #중복 제거 완화
        
    items = [(tok, c) for tok, c in tf.items() if c >= min_tf]
    items.sort(key=lambda x: (-x[1], x[0])) 
    id2token = [tok for tok, _ in items]
    token2id = {t:i for i,t in enumerate(id2token)}
    return token2id, id2token


# 2. 윈도우 슬라이딩: (center, context) 페어 누적

def build_window_pairs(corpus_tokens, token2id, window=2, *, 
                       distance_weight=False, undirected=True):
    """
    window: ±window 범위
    distance_weight=True일 경우 1/거리 가중치, False일 경우 1씩 가산
    directed=False  무방향: 오른쪽만 스캔하여 (i<j) 쌍 1회 누적
    directed=True   방향: 좌우 모두 스캔하여 전부 누적
    """
    pair_counts = Counter()
    for sent in corpus_tokens:
        ids = [token2id[t] for t in sent if t in token2id] #토큰 리스트를 인덱스 리스트로 전환
        n = len(ids)
        for i in range(n):
            ci = ids[i]
            
            if undirected:
                # 무방향: 오른쪽만 스캔 (i < j) 
                # 중복 방지 과정인데, 윈도우를 오른쪽으로만 펼친다고 보면 좋음
                hi = min(n - 1, i + window)
                # 오른쪽 벽. 윈도우가 문장 끝에 다다랐을 때 처리 방식
                for j in range(i + 1, hi + 1):
                    cj = ids[j]
                    w = (1.0 / (j - i)) if distance_weight else 1.0
                    # 거리 웨이트 옵션 / 통상 1이 들어간다고 생각
                    u, v = (ci, cj) if ci < cj else (cj, ci)  # 인덱스 정규화
                    pair_counts[(u, v)] += w
                    
            else:
                lo, hi = max(0, i - window), min(n - 1, i + window)
                # 왼쪽 오른쪽 벽, 즉 윈도우가 문장 한계에 부딪힐 때 고려
                for j in range(lo, hi + 1):
                    if j == i: 
                        continue
                    cj = ids[j]
                    w = (1.0 / abs(i - j)) if distance_weight else 1.0
                    # 거리 웨이트 옵션 / 통상 1이 들어간다고 생각
                    pair_counts[(ci, cj)] += w  # 방향 유지

    return pair_counts


# 2. 페어 카운트를 희소행렬로 변환

def pairs_to_sparse(pair_counts, vocab_size, undirected=True, zero_diagonal=True):
    """
    pair_counts: Counter[(u,v)] (u!=v 가정)
    vocab_size: |V|
    undirected=True면 대칭 그래프 가정
    """
    rows, cols, data = [], [], []
    for (u, v), w in pair_counts.items():
        rows.append(u); cols.append(v); data.append(w)
        if undirected:
            rows.append(v); cols.append(u); data.append(w)

    W = sp.coo_matrix((data, (rows, cols)), shape=(vocab_size, vocab_size)).tocsr()
    #csr을 쓰면 좋다고 함 
    
    if zero_diagonal:
        W = W.tolil()
        W.setdiag(0)
        W = W.tocsr()
        W.eliminate_zeros()
    return W


# 3. 희소행렬 → NetworkX (임계치로 weight 작은 link 제거 후 상삼각만 사용)
def sparse_to_networkx(
    W,
    id2token,
    weight_threshold: float = 0.0,
    directed: bool = False,
    use_upper_if_undirected: bool = True,
    remove_self_loops: bool = False,
):
    """
    W: V×V 희소행렬(CSR 권장). 가중치는 'weight'라는 노드 속성으로.
    directed=False  -> 무방향 Graph
    directed=True   -> 방향 DiGraph (비대칭 행렬 허용)

    use_upper_if_undirected=True면 무방향에서 상삼각만 전달(중복 방지).
    remove_self_loops=True면 자기루프 제거.
    """
    W = W.tocsr().copy()

    # 임계치 미만 weight link 제거 (networkx 그래프 만들기 전에 0으로)
    if weight_threshold > 0:
        mask = W.data >= weight_threshold
        W.data[~mask] = 0.0
        W.eliminate_zeros()

    # 자기루프 제거 옵션
    # 이 맥락에서는 윈도우 안에 중심 단어와 동일한 단어가 등장했을 경우 처리 판단
    if remove_self_loops:
        W = W.tolil()
        W.setdiag(0)
        W = W.tocsr()
        W.eliminate_zeros()

    if directed:
        # 방향 그래프: 비대칭 OK, 그대로 전달
        G = nx.from_scipy_sparse_array(W, create_using=nx.DiGraph, edge_attribute="weight")
    else:
        # 무방향 그래프: 상삼각만 전달(중복/가중치 합산 혼선 방지)
        W_use = sp.triu(W, k=1).tocsr() if use_upper_if_undirected else W
        G = nx.from_scipy_sparse_array(W_use, create_using=nx.Graph, edge_attribute="weight")

    # 정수 노드를 토큰 라벨로
    nx.relabel_nodes(G, {i: tok for i, tok in enumerate(id2token)}, copy=False)
    return G
```

앞서 우리가 했던 절차와 크게 다르지 않습니다. 사실 앞에서 정의했던 사전은 그대로 써도 되는데, 이렇게 간단하게 운용할 수도 있다는 점 알려드리기 위해 소략한 버전을 제시해봤습니다. 

위 함수들에서 아예 처음 보는 구조의 함수는 사실 build_window_pairs 밖에 없습니다. 천천히 보시면 알겠지만, 그냥 for 구문안에서 인덱스가 움직이면서 그 인덱스가 가리키는 토큰과 특정 window 차이 안에 있는 인덱스의 토큰을 연결되었다고 기록하는 장치입니다. 별로 복잡할게 없죠. 

#### 3.3.2. 네트워크 만들고 시각화

```python
token2id, id2token = build_vocab(corpus_tokens, min_tf=1)

# 1) 윈도우=2, 빈도 카운트 방식(중복 유지)
pair_counts = build_window_pairs(
    corpus_tokens, token2id, window=2, distance_weight=False, undirected=True
)

# 2) 희소행렬
W_win = pairs_to_sparse(pair_counts, vocab_size=len(id2token), undirected=True, zero_diagonal=True)

# 3) 네트워크
G_win = sparse_to_networkx(W_win, id2token, weight_threshold = 1.0)

print(G_win.nodes(), "\n")
print(G_win.edges(data = True), "\n")
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 8))

# spring layout
pos = nx.spring_layout(G_win, seed=42) 

# 네트워크 시각화는 반드시 노드 배치의 논리를 결정하는 레이아웃이 필요합니다

# 기본 그리기 (노드 + 엣지 + 레이블)
nx.draw(G_win, pos,
        with_labels=True,
        font_family="Malgun Gothic",  # 한글 노드 깨짐 방지
        node_color="lightblue",
        node_size=100,
        font_size=8,
        edge_color="gray",
        alpha=0.8)

plt.title("Window based co-occurrence Network (spring layout)")
plt.show()
```

앞서 본 것과 약간 그림이 다르죠? 지금은 자료의 문장 길이가 짧아서 차이가 덜 두드러지는데, 문장 길이가 길어지면 당연히 '문장 기반 동시 출현 네트워크'와 '윈도우 기반 근접 출현 네트워크'는 그 형태가 아주 달라질 것입니다.
